# Linear Transformation

In this notebook, we will explore the influence that a $2\times2$ matrix has on a vector.

As mentioned in the seminar, I really recommend watching https://www.youtube.com/watch?v=fNk_zzaMoSs&list=PLZHQObOWTQDPD3MizzM2xVFitgF8hE_ab

```python
# for plotting
import pylab as plt
# for matrix stuff
import numpy as np
```

At first, we create a set of $N_l$ lines in x- and y-direction.

```python
# points per line
N = 100
# lines per direction
Nl = 11
line_set = np.zeros((2, Nl*2, 100))

line_set[0, :Nl] = np.array([N*[i] for i in np.linspace(-5, 5, Nl)])
line_set[0, Nl:] = np.array([np.linspace(-5, 5, N) for i in np.linspace(-5, 5, Nl)])
line_set[1, Nl:] = np.array([N*[i] for i in np.linspace(-5, 5, Nl)])
line_set[1, :Nl] = np.array([np.linspace(-5, 5, N) for i in np.linspace(-5, 5, Nl)])
```

```python
print(line_set.shape)
```

As we can see, `line_set` is of dimension $(2, 22, 100)$, since we have $2$ coordinates for $2\cdot N_l = 22$ lines, which are sampled by $100$ points each.

```python
plt.figure(figsize=(5, 5))
for i in range(2*Nl):
    plt.plot(line_set[0, i], line_set[1, i], 'k-')
```

If we apply 
$$M = \begin{pmatrix}
1 & 1\\
0 & 1
\end{pmatrix}$$
on our set of lines, $\boldsymbol{e}_x$ is mapped to itself and $\boldsymbol{e}_y$ to $(1, 1)^T$, i.e., M is stretching the grid in y-direction.

```python
M1 = np.array([[1, 1],
               [0, 1]])
l1 = np.tensordot(M1, line_set, axes=(1, 0))
```

```python
plt.figure(figsize=(5, 5))
for i in range(2*Nl):
    plt.plot(line_set[0, i], line_set[1, i], 'k-')
    plt.plot(l1[0, i], l1[1, i], 'r-', zorder=100)
```

Another example would be
$$M = \begin{pmatrix}
0.5 & -0.5\\
0.5 & 0.5
\end{pmatrix}$$
Now, $\boldsymbol{e}_x$ is mapped to $(0.5, 0.5)^T$ and $\boldsymbol{e}_y$ to $(0.5, -0.5)^T$, leading to a $\pi/4$ rotation and a compression of the grid. The area of each square will change according to $\operatorname{det}(M)=0.5$.

```python
M2 = np.array([[0.5, -0.5],
               [0.5,  0.5]])
l2 = np.tensordot(M2, line_set, axes=(1, 0))
```

```python
plt.figure(figsize=(5, 5))
for i in range(2*Nl):
    plt.plot(line_set[0, i], line_set[1, i], 'k-')
    plt.plot(l2[0, i], l2[1, i], 'r-', zorder=100)
```
